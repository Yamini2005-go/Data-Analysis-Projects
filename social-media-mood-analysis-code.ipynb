```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/Response_social (2).csv")
df.columns = df.columns.str.strip() # Strip whitespace from column names
print(df)
sns.countplot(data = df, x = "overall_mood_impact")
plt.title("Mood Distribution of the users")
plt.show()

age_midpt = {
    "under 20" : 15,
    "20-24" : 22,
    "25-30" : 27.5,
    "30+" : 35 # Corrected key to match dataframe
}

df["age_midpt"] = df["age"].map(age_midpt)
sns.scatterplot(data = df, x = "age_midpt", y = "avg_time_scale")
plt.title("Average time on socialmedia VS age")
plt.show()
```
